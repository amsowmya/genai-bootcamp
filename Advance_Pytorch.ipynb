```python
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear1 = nn.Linear(num_features, 3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
    return out
```

# Sequential method

```python
import torch
import torch.nn as nn

class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()

    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, features):
    out = self.network(features)
    return out
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, 1:], data.iloc[:, 0], test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(X_train)
x_test = scaler.transform(X_test)

encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
x_train_tensor = torch.from_numpy(x_train.astype(np.float32))
x_test_tensor = torch.from_numpy(x_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
class MySimpleNN(nn.Module):
  def __init__(self, num_features):
    super().__init__()

    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, features):
    out = self.network(features)
    return out
```

```python
data.shape
```

```python
x_train.shape
```

```python
learning_rate = 0.1
epochs = 100
```

```python
loss_function = nn.BCELoss()
```

```python
model = MySimpleNN(x_train.shape[1])
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
for epoch in range(epochs):
  y_pred = model(x_train_tensor)
  loss = loss_function(y_pred, y_train_tensor.view(-1, 1))

  # clear gradient
  optimizer.zero_grad()

  loss.backward()
  # parameters update
  optimizer.step()

  print(f"Epoch: {epoch + 1}, Loss: {loss.item()}")
```

```python
with torch.no_grad():
  y_pred = model(x_test_tensor)

  # y_pred_label = torch.where(y_pred > 0.5, 1, 0)
  y_pred_label = (y_pred > 0.5).float()
  accuracy = (y_pred_label == y_test_tensor).float().mean()
  print(f"Accuracy: {accuracy.item()}")
```

# Dataset and DataLoader

```python
from sklearn.datasets import make_classification
import torch
```

```python
X, y = make_classification(
    n_samples=10,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42
)
```

```python
X
```

```python
y
```

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)
```

```python
X
```

```python
y
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataSet(Dataset):
  def __init__(self, features, labels):
    self.features = features
    self.labels = labels

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self, idx):
    return self.features[idx], self.labels[idx]
```

```python
dataset = CustomDataSet(X, y)
```

```python
len(dataset)
```

```python
dataset[1]
```

```python
dataset[7]
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=False)
```
